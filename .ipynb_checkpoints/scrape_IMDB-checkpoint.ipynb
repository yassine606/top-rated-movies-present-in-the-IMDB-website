```python
import requests 
```

```python
from bs4 import BeautifulSoup
```

```python
url = "https://www.imdb.com/chart/top"
page = requests.get(url)
page.status_code
```

```python
page_content = page.text
soup = BeautifulSoup(page_content, "html.parser")
```

```python
movie_list=soup.find('tbody',{'class':'lister-list'}).find_all('tr')
```

```python
#extraire les noms des films 
for movie in movie_list : 
    name=movie.find('td',{'class':'titleColumn'}).a.text
    print(name)
    
    
#for movie in movie_list :  
 #   year_film.append(movie.find('td',{'class':'titleColumn'}).span.text)
  #  print(year_film)
    
    
#for movie in movie_list : 
 #   rating_film.append(movie.find('td',{'class':'ratingColumn imdbRating'}).strong.text)
  #  print(rating_film)
    
#print(film_names)
#print('\n')
#print(year_film)
#print('\n')
#print(rating_film)
```

```python
#extraire les ranks 
for movie in movie_list : 
    rank=movie.find('td',{'class':'titleColumn'}).get_text().split('.')[0]
    print(rank)
```

```python
#extraire 'year' des films 
for movie in movie_list : 
    year=movie.find('td',{'class':'titleColumn'}).span.text[1:-1]
    print(year)
```

```python
#extraire les rating 
for movie in movie_list : 
    year=movie.find('td',{'class':'ratingColumn imdbRating'}).strong.text
    print(year)
    
```

```python
from csv import writer
```

```python
with open('top_rating.csv', 'w', encoding='utf8', newline='') as f:
    thewriter = writer(f)
    header = ['Rank', 'Name', 'year', 'Rating']
    thewriter.writerow(header)
    
    
    for movie in movie_list : 
        
    
        rank=movie.find('td',{'class':'titleColumn'}).get_text().split('.')[0]
        name=movie.find('td',{'class':'titleColumn'}).a.text
        year=movie.find('td',{'class':'titleColumn'}).span.text[1:-1]
        rating=movie.find('td',{'class':'ratingColumn imdbRating'}).strong.text
    
    
        info=[rank,name,year,rating]
        thewriter.writerow(info)
```
